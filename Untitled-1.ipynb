```python
# Import necessary libraries
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
import torchvision.transforms as transforms
from torchvision.datasets import ImageFolder
import matplotlib.pyplot as plt

# Define CapsNet components
class PrimaryCaps(nn.Module):
    def __init__(self, in_channels, out_channels, capsule_dim):
        super(PrimaryCaps, self).__init__()
        self.conv = nn.Conv2d(in_channels, out_channels, kernel_size=9, stride=2)
        self.capsule_dim = capsule_dim

    def forward(self, x):
        x = self.conv(x)
        batch_size = x.size(0)
        x = x.view(batch_size, -1, self.capsule_dim)  # [batch, num_capsules, dim]
        return squash(x)

def squash(vectors):
    """Squashing function for capsule vectors."""
    squared_norm = (vectors ** 2).sum(dim=-1, keepdim=True)
    scale = squared_norm / (1 + squared_norm) / torch.sqrt(squared_norm + 1e-7)
    return scale * vectors

class DigitCaps(nn.Module):
    def __init__(self, num_capsules, num_routes, out_channels):
        super(DigitCaps, self).__init__()
        self.num_capsules = num_capsules
        self.num_routes = num_routes
        self.out_channels = out_channels
        self.W = nn.Parameter(torch.randn(1, num_routes, num_capsules, out_channels, 8))

    def forward(self, x):
        batch_size = x.size(0)
        x = x.unsqueeze(2)  # [batch, num_routes, 1, in_channels]
        W = self.W.expand(batch_size, -1, -1, -1, -1)
        u_hat = torch.matmul(W, x.unsqueeze(-1))  # [batch, num_routes, num_caps, out_channels, 1]

        # Simplified routing (3 iterations)
        b = torch.zeros(batch_size, self.num_routes, self.num_capsules, 1, 1).to(x.device)
        for _ in range(3):  # Routing iterations
            c = torch.softmax(b, dim=2)  # Coupling coefficients
            s = (c * u_hat).sum(dim=1, keepdim=True)  # Weighted sum
            v = squash(s.squeeze(1))  # [batch, num_caps, out_channels, 1]
            b = b + (u_hat * v.unsqueeze(1)).sum(dim=-1, keepdim=True)  # Update couplings
        return v.squeeze(-1)  # [batch, num_caps, out_channels]

class CapsNet(nn.Module):
    def __init__(self, num_classes):
        super(CapsNet, self).__init__()
        self.conv1 = nn.Conv2d(3, 256, kernel_size=9, stride=1)
        self.primary_caps = PrimaryCaps(256, 32 * 8, 8)  # 32 capsules, 8D each
        self.digit_caps = DigitCaps(num_classes, 32 * 6 * 6, 16)  # num_classes capsules

    def forward(self, x):
        x = torch.relu(self.conv1(x))
        x = self.primary_caps(x)
        x = self.digit_caps(x)
        return x

# Margin Loss for CapsNet
def margin_loss(outputs, labels, m_plus=0.9, m_minus=0.1, lambda_=0.5):
    batch_size = outputs.size(0)
    v_length = torch.sqrt((outputs ** 2).sum(dim=-1))  # [batch, num_caps]
    
    # One-hot encode labels
    one_hot = torch.zeros(batch_size, outputs.size(1)).to(outputs.device)
    one_hot.scatter_(1, labels.view(-1, 1), 1)
    
    # Margin loss terms
    T = one_hot  # Target (1 for correct class, 0 otherwise)
    loss_pos = T * torch.relu(m_plus - v_length) ** 2  # Positive class loss
    loss_neg = (1 - T) * torch.relu(v_length - m_minus) ** 2  # Negative class loss
    loss = loss_pos + lambda_ * loss_neg
    return loss.sum(dim=1).mean()

# Hyperparameters
data_dir = "/home/idrone2/Desktop/Cashew"
num_classes = 5
epochs = 20
batch_size = 32
learning_rate = 0.001
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Data loading
transform = transforms.Compose([
    transforms.Resize((28, 28)),  # Adjust based on your data
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

dataset = ImageFolder(root=data_dir, transform=transform)
train_loader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

# Initialize model, optimizer
model = CapsNet(num_classes).to(device)
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

# Lists to store metrics
train_losses = []
train_accuracies = []

# Training loop with accuracy and graphing
for epoch in range(epochs):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0
    
    for i, (images, labels) in enumerate(train_loader):
        images, labels = images.to(device), labels.to(device)

        # Forward pass
        optimizer.zero_grad()
        outputs = model(images)
        loss = margin_loss(outputs, labels)

        # Backward pass
        loss.backward()
        optimizer.step()

        # Stats
        running_loss += loss.item()
        v_length = torch.sqrt((outputs ** 2).sum(dim=-1))  # Capsule lengths
        _, predicted = torch.max(v_length.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

        if i % 10 == 9:  # Print every 10 batches
            avg_loss = running_loss / 10
            accuracy = 100 * correct / total
            print(f"Epoch [{epoch+1}/{epochs}], Step [{i+1}/{len(train_loader)}], "
                  f"Loss: {avg_loss:.4f}, Accuracy: {accuracy:.2f}%")
            running_loss = 0.0
            correct = 0
            total = 0

    # Store epoch-wise metrics
    epoch_loss = running_loss / len(train_loader) if running_loss > 0 else loss.item()
    epoch_accuracy = 100 * correct / total if total > 0 else 100 * (predicted == labels).sum().item() / labels.size(0)
    train_losses.append(epoch_loss)
    train_accuracies.append(epoch_accuracy)
    print(f"Epoch [{epoch+1}/{epochs}] completed, Avg Loss: {epoch_loss:.4f}, Avg Accuracy: {epoch_accuracy:.2f}%")

# Plotting graphs
plt.figure(figsize=(12, 5))

# Plot Loss
plt.subplot(1, 2, 1)
plt.plot(range(1, epochs + 1), train_losses, label='Training Loss', color='blue')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss Over Time')
plt.legend()
plt.grid(True)

# Plot Accuracy
plt.subplot(1, 2, 2)
plt.plot(range(1, epochs + 1), train_accuracies, label='Training Accuracy', color='green')
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.title('Training Accuracy Over Time')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

print("Training finished!")
```
