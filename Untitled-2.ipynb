```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to convert RGB to HSI
def rgb_to_hsi(image):
    image = image.astype(np.float32) / 255.0
    r, g, b = cv2.split(image)
    intensity = (r + g + b) / 3.0
    
    min_val = np.minimum(np.minimum(r, g), b)
    saturation = 1 - (3 / (r + g + b + 1e-6)) * min_val
    
    num = 0.5 * ((r - g) + (r - b))
    den = np.sqrt((r - g) ** 2 + (r - b) * (g - b)) + 1e-6
    theta = np.arccos(num / den)
    hue = np.where(b > g, 2 * np.pi - theta, theta)
    hue = hue / (2 * np.pi)  # Normalize to [0,1]
    
    hsi = cv2.merge((hue, saturation, intensity))
    return (hsi * 255).astype(np.uint8)

# Function to convert RGB to HSL
def rgb_to_hsl(image):
    image = image.astype(np.float32) / 255.0
    r, g, b = cv2.split(image)
    max_val = np.maximum(np.maximum(r, g), b)
    min_val = np.minimum(np.minimum(r, g), b)
    
    lightness = (max_val + min_val) / 2.0
    
    delta = max_val - min_val
    saturation = np.where(lightness > 0, delta / (1 - np.abs(2 * lightness - 1) + 1e-6), 0)
    
    num = ((max_val - g) + (max_val - b)) / 2
    den = np.sqrt((r - g) ** 2 + (r - b) * (g - b)) + 1e-6
    theta = np.arccos(num / den)
    hue = np.where(b > g, 2 * np.pi - theta, theta)
    hue = hue / (2 * np.pi)  # Normalize to [0,1]
    
    hsl = cv2.merge((hue, saturation, lightness))
    return (hsl * 255).astype(np.uint8)

# Load the image
image = cv2.imread('/home/idrone2/Desktop/RSM_classification/RSM_Modrate/RSM_00706.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Convert to different color spaces
hsv_image = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
hsi_image = rgb_to_hsi(image)
hsl_image = rgb_to_hsl(image)

# Plot the images
fig, axes = plt.subplots(1, 4, figsize=(24, 8))
axes[0].imshow(image)
axes[0].set_title("Original RGB Image")
axes[1].imshow(hsv_image)
axes[1].set_title("HSV Image")
axes[2].imshow(hsi_image)
axes[2].set_title("HSI Image")
axes[3].imshow(hsl_image)
axes[3].set_title("HSL Image")

for ax in axes:
    ax.axis("off")

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to convert RGB to HSI
def rgb_to_hsi(image):
    image = image.astype(np.float32) / 255.0
    r, g, b = cv2.split(image)
    intensity = (r + g + b) / 3.0
    
    min_val = np.minimum(np.minimum(r, g), b)
    saturation = 1 - (3 / (r + g + b + 1e-6)) * min_val
    
    num = 0.5 * ((r - g) + (r - b))
    den = np.sqrt((r - g) ** 2 + (r - b) * (g - b)) + 1e-6
    theta = np.arccos(num / den)
    hue = np.where(b > g, 2 * np.pi - theta, theta)
    hue = hue / (2 * np.pi)  # Normalize to [0,1]
    
    hsi = cv2.merge((hue, saturation, intensity))
    return (hsi * 255).astype(np.uint8)

# Function to convert RGB to HSL
def rgb_to_hsl(image):
    image = image.astype(np.float32) / 255.0
    r, g, b = cv2.split(image)
    max_val = np.maximum(np.maximum(r, g), b)
    min_val = np.minimum(np.minimum(r, g), b)
    
    lightness = (max_val + min_val) / 2.0
    
    delta = max_val - min_val
    saturation = np.where(lightness > 0, delta / (1 - np.abs(2 * lightness - 1) + 1e-6), 0)
    
    num = ((max_val - g) + (max_val - b)) / 2
    den = np.sqrt((r - g) ** 2 + (r - b) * (g - b)) + 1e-6
    theta = np.arccos(num / den)
    hue = np.where(b > g, 2 * np.pi - theta, theta)
    hue = hue / (2 * np.pi)  # Normalize to [0,1]
    
    hsl = cv2.merge((hue, saturation, lightness))
    return (hsl * 255).astype(np.uint8)

# Load the image
image = cv2.imread('/home/idrone2/Desktop/RSM_classification/RSM_Severe/RSM_00488.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Convert to different color spaces
hsv_image = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
hsi_image = rgb_to_hsi(image)
hsl_image = rgb_to_hsl(image)

# Step 2: Creating a Mask for Red Color
lower_red1 = np.array([0, 120, 70])
upper_red1 = np.array([10, 255, 255])
lower_red2 = np.array([170, 120, 70])
upper_red2 = np.array([180, 255, 255])

mask1 = cv2.inRange(hsv_image, lower_red1, upper_red1)
mask2 = cv2.inRange(hsv_image, lower_red2, upper_red2)
red_mask = mask1 + mask2

# Step 3: Applying Contour Detection
contours, _ = cv2.findContours(red_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
detected_image = image.copy()
cv2.drawContours(detected_image, contours, -1, (0, 255, 0), 2)

# Step 4: Refining the Detection Results
for contour in contours:
    if cv2.contourArea(contour) > 500:  # Filter small areas
        x, y, w, h = cv2.boundingRect(contour)
        cv2.rectangle(detected_image, (x, y), (x + w, y + h), (255, 0, 0), 2)

# Plot the images
fig, axes = plt.subplots(1, 4, figsize=(16, 4))
axes[0].imshow(image)
axes[0].set_title("Original RGB Image")
axes[1].imshow(hsv_image)
axes[1].set_title("HSV Image")
axes[2].imshow(red_mask, cmap='gray')
axes[2].set_title("Red Color Mask")
axes[3].imshow(detected_image)
axes[3].set_title("Detected Contours")

for ax in axes:
    ax.axis("off")

plt.show()
```
